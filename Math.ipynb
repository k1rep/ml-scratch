```python
from scipy import linalg
A = np.array([[3,4,5],
              [5,6,7],
              [3,5,8]])
y = np.array([8,9,10])
A_in = linalg.inv(A)
print(A_in)
print("*"*100)
print(np.dot(A_in,y))
```

```python
A = np.array([[1,1],
              [2,1],
              [1,1]])
b = np.array([0,0,2])

A_T_A = np.dot(A.T,A)
A_T_A_in = linalg.inv(A_T_A)
x_hat = np.dot(np.dot(A_T_A_in,A.T),b)
p = np.dot(A,x_hat)
p_matrix = np.dot(A,np.dot(A_T_A_in,A.T))
print("x_hat:",x_hat)
print("p:",p)
print("p_matrix:",p_matrix)
```

```python
# theorem of large numbers
import random
import matplotlib.pyplot as plt
def coin_flip(min, max):
    ratios = []
    x = range(min, max+1)
    for number_flips in x:
        heads = 0
        for i in range(number_flips):
            if random.random() < 0.5:
                heads += 1
        tails = number_flips - heads
        ratios.append(heads/float(tails))
        
    plt.title("Heads/Tails Ratios")
    plt.xlabel("Number of Flips")
    plt.ylabel("Heads/Tails")
    plt.plot(x, ratios)
    plt.show()
coin_flip(2, 10000)
```

```python
random_data = np.random.randint(1,7,60000)
print(random_data)
print('mean:',np.mean(random_data))
print('variance:',np.var(random_data))
plt.hist(random_data, bins=50, density=False)
plt.show()
```

```python
sample_100 = []
for i in range(0,100):
    sample_100.append(random_data[int(np.random.random()*len(random_data))])
print(sample_100)
print('mean:',np.mean(sample_100))
print('variance:',np.var(sample_100))
```

```python
# central limit theorem
sample_100_mean = []
for i in range(0,50000):
    sample_100 = []
    for i in range(0,100):
        sample_100.append(random_data[int(np.random.random()*len(random_data))])
    sample_100_mean.append(np.mean(sample_100))
    samples_many_mean = np.mean(sample_100_mean)
    
print("mean:", samples_many_mean)
plt.hist(sample_100_mean, bins=200, density=False)
plt.show()
```
